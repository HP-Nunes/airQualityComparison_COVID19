# Introduction

<br>
Particulate Matter is a mixture of solid particles and liquid droplets in the air of varied size and shape: PM10—coarse particulate matter, PM2.5—fine particulate matter, and Ultrafine particles. While PM10 (dust or smoke) may be irritating to the mucous membranes, PM2.5 poses a significant health risk because they are commonly inhaled into the lungs and bloodstream, causing serious respiratory health issues and posing a significant health risks for people with asthma and heart conditions. In order to systematically assess the exposure of the general population to the hazards of air pollution, air quality monitoring is required, often under a regulatory mandate (Borghi et al, 2017).

According to the American Lung Association's State of the Air Report for 2019, the San Jose-San Francisco-Oakland metropolitan area ranks as high as third out of 216 metropolitan areas for 24-hour particle pollution and fifth for annual particle pollution out of 204 metropolitan areas.
(http://www.stateoftheair.org/city-rankings/msas/san-jose-san-francisco-oakland-ca.html#pmann)

The purpose of this project is to explore whether 'shelter-in-place' orders due to COVID-19 for the city of San Francisco has had a noticeable impact on air quality. This is based on the presumption that suspended economic activities and a population instructed to self-quarantine has had the effect of reducing activities which contribute to fine particle pollution. Such activities include vehicular traffic, construction, and manufacturing. 

The parameters of this research are as follow:
* A focus on 24-hour PM2.5 particle measurements (measured in µg/m3);
* An overall temporal scope from January 1st - April 30th;
* The confinement period is defined from March 17th to April 30th;
* A baseline comparison with 2019 and 2018 measurements for that same period of measurements;

(https://abc7news.com/timeline-of-coronavirus-us-coronvirus-bay-area-sf/6047519/)
